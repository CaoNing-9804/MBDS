```python
k1=100/60
k2=600/60
k3=150/60
def fe(e,s,es,p):
    ce=-k1*e*s+k2*es+k3*es
    return ce

def fs(e,s,es,p):
    cs=-k1*e*s+k2*es
    return cs

def fes(e,s,es,p):
    ces=k1*e*s-k2*es-k3*es
    return ces

def fp(e,s,es,p):
    cp=k3*es
    return cp
```

```python
earray=[]
sarray=[]
esarray=[]
parray=[]
array=[]
```

```python
def RK4():
    h=0.01
    a=0
    e0=1
    s0=10
    es0=0
    p0=0
    while a<=1:
        #print(a)
        array.append(a)
        earray.append(e0)
        sarray.append(s0)
        esarray.append(es0)
        parray.append(p0)
        a+=h

        e1=fe(e0,s0,es0,p0)
        #print('ok')
        te1=e0+e1*h/2
        s1=fs(e0,s0,es0,p0)
        ts1=s0+s1*h/2
        es1=fes(e0,s0,es0,p0)
        tes1=es0+es1*h/2
        p1=fp(e0,s0,es0,p0)
        tp1=p0+p1*h/2
        #print(e1,s1,es1,p1)
        ################step1##################

        e2=fe(te1,ts1,tes1,tp1) 
        te2=e0+e2*h/2
        s2=fs(te1,ts1,tes1,tp1)
        ts2=s0+s2*h/2
        es2=fes(te1,ts1,tes1,tp1)
        tes2=es0+es2*h/2
        p2=fp(te1,ts1,tes1,tp1)
        tp2=p0+p2*h/2
        #print(e2,s2,es2,p2)
        ################step2##################

        e3=fe(te2,ts2,tes2,tp2) 
        te3=e0+e3*h/2
        s3=fs(te2,ts2,tes2,tp2)
        ts3=s0+s3*h/2
        es3=fes(te2,ts2,tes2,tp2)
        tes3=es0+es3*h/2
        p3=fp(te2,ts2,tes2,tp2)
        tp3=p0+p3*h/2
        #print(e3,s3,es3,p3)
        ################step3##################

        e4=fe(te3,ts3,tes3,tp3) 
        s4=fs(te3,ts3,tes3,tp3)
        es4=fes(te3,ts3,tes3,tp3)
        p4=fp(te3,ts3,tes3,tp3)
        #print(e4,s4,es4,p4)
        ################step4##################
        
        

        e0=e0+(e1+2*e2+2*e3+e4)*h/6
        s0=s0+(s1+2*s2+2*s3+s4)*h/6
        es0=es0+(es1+2*es2+2*es3+es4)*h/6
        p0=p0+(p1+2*p2+2*p3+p4)*h/6
        print(e0,s0,es0,p0)
```
